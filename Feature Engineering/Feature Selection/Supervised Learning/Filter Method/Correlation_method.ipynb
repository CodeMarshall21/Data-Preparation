```python
import pandas as pd
```

```python
data = {
    "Name" : ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hannah", "Ian"],
    "Age" : [25, 30, 35, 45, 28, 40, 30, 25, 50],
    "Salary" : [50000, 300000, 400000, 500000, 85000, 600000, 300000, 52000, 600000],
    "City" : ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix", "Philadelphia", "San Antonio", "San Diego", "Dallas"],
    "Weight" : [130, 180, 175, 200, 140, 190, 160, 135, 210],
    "Loan Status": ["Denied", "Denied", "Approved", "Approved", "Denied", "Approved", "Denied", "Denied", "Denied"]
}

df = pd.DataFrame(data)
df
```

```python
df["Loan Status"] = (df["Loan Status"] == "Approved").astype(int)
df
```

```python
df[["Age", "Salary", "Weight", "Loan Status"]].corr()
```

```python
correlation = df[["Age", "Salary", "Weight", "Loan Status"]].corr()["Loan Status"]
correlation
```

```python
correlation = correlation.abs()
correlation
```

```python
sorted_features = correlation.sort_values(ascending = False)
sorted_features
```
