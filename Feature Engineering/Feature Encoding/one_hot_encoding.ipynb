```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
data = {
    'Color': ['Red', 'Blue', 'Green', 'Blue', 'Red'],
    'Size': ['S', 'M', 'L', 'XL', 'M'],
}

df = pd.DataFrame(data)
```

```python
encoder = OneHotEncoder()
```

```python
encoded_color = encoder.fit_transform(df[["Color"]]).toarray()
```

```python
encoded_color = pd.DataFrame(encoded_color, columns = encoder.get_feature_names_out(["Color"]))
encoded_color
```

```python
df_encoded = pd.concat([df["Color"], encoded_color], axis=1)
df_encoded
```
